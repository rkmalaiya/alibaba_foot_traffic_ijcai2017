### Splitting the data into training, testing, and validation sets

We'll save the last 21 days of the data to use as a test set after we've trained the network. We'll use this set to make predictions and compare them with the actual number of riders.

```python
data = pd.read_pickle("cff_15dayperiod.pkl")
data.shape
```

```python
import tensorflow as tf
input_nodes_num = 15
output_nodes_num = 1

tf_x = tf.placeholder(tf.float32, [None, input_nodes_num])
tf_W = tf.Variable(tf.zeros([input_nodes_num, output_nodes_num]))
tf_b = tf.Variable(tf.zeros([output_nodes_num]))

tf_y = tf.nn.softmax(tf.matmul(tf_x, tf_W) + tf_b)
tf_y_ = tf.placeholder(tf.float32, [None, output_nodes_num])
```

```python
cross_entropy = tf.reduce_mean(
                                -tf.reduce_sum(tf_y_ * tf.log(tf_y), 
                                reduction_indices=[1]))
```

```python
train_step = tf.train.GradientDescentOptimizer(0.5).minimize(cross_entropy)
```
